**Function to import dataset**

```python
def import_dataset(file):
  df = pd.read_excel(file, header=None)
  return df
```

**Function to extract one example**

```python
def extract_example(df, example_no):
  i = example_no
  column1 = (i*2)
  column2 = column1 + 1
  dataset = df.iloc[:,column1:column2+1]

  return dataset
```

**Plot datapoints**

```python
def plot(example, i):
  plt.scatter(example[i*2], example[(i*2)+1], c = 'r', marker = '.')
```

**Compute Covariance Matrix**

```python
def covariance(example):
  cov_matrix = np.cov(example.T)
  cov = cov_matrix[0][1]
  varA = cov_matrix[0][0]
  varB = cov_matrix[1][1]
  var = [varA, varB]
  return [cov_matrix, cov, var]
```

# **Main Function**

```python
df = import_dataset('/content/multNormal.xlsx')

cov_array = []
var_array = []
no_of_example = []
no_of_example_var = []

plt.figure(figsize=(10,10))

#For every example
for i in range(10):
  example = extract_example(df, i)
  plt.subplot(2, 5, i+1)
  plt.xlim([-20,20])
  plt.ylim([-10,10])
  plot(example,i)
  [cov_matrix, cov, var] = covariance(example)

  print("\nExample No:", i+1)
  print(cov_matrix)
  
  cov_array.append(cov)
  var_array.append(var)
  no_of_example.append(i+1)
  no_of_example_var.append([i+1,i+1])

plt.show()
```

```python
#Display cov plot
print("Covariance vs Example Plot")
plt.scatter(no_of_example, cov_array)
plt.show()

#Display var plot
print("Variance vs Example Plot")
plt.scatter(no_of_example_var, var_array)
plt.show()
```
